## The functions in this file were used to perform stochastic simulations, test stability of the model, and create corresponding plots. These were primarily designed for our own insight, and therefore this file does not include extended explanation.

```python
#simulation results

def decomlist(result):
    mean = []
    std = []

    for i in range(len(result[0])):
        list = []    
        for j in range(len(result)):
            list.append(result[j][i])
    
        #print('now',datalist[j][i])
        mean = np.append(mean,np.mean(list))
        std = np.append(std, np.std(list, ddof=1))
    confi_u=mean+std
    confi_d=mean-std
    
    return mean,confi_u, confi_d
```

```python
#simulation 50  iterations 60,  random setting 0.2, 0.5, 0.3

def CD(names, S):
    simulations = 100
    steps = 30

    drawing = MesaAxelrodNetwork(final_network, S,0.2, 0.5, 0.3)

    datalist = []
    for i in range(simulations):    
        drawing = MesaAxelrodNetwork(final_network, S,0.2, 0.5, 0.3)

        for j in range(steps):
            drawing.step()
            save = drawing.datacollector.get_model_vars_dataframe()[names][1:]/(211)
        datalist.append(save.tolist())
        
    mean, up,lo = decomlist(datalist)
    plt.plot(range(len(mean)),mean,'-',color = "gold",label=names)
    plt.fill_between( range(len(mean)),up,lo,color="wheat",alpha=0.5,label="std")  
    plt.xlabel('Iterations')
    plt.ylabel('Percentage')    
    
    
def stra(names, S):
    simulations = 50
    steps = 60

    drawing = MesaAxelrodNetwork(final_network, S, 0.2, 0.5, 0.3)

    datalist = []
    for i in range(simulations):    
        drawing = MesaAxelrodNetwork(final_network, S,0.2, 0.5, 0.3)

        for j in range(steps):
            drawing.step()
            save = drawing.datacollector.get_model_vars_dataframe()[names]/(10.55)  
        datalist.append(save.tolist())
        
    mean, up,lo = decomlist(datalist)
    plt.plot(range(len(mean)),mean,'-',label=names)
    plt.fill_between( range(len(mean)),up,lo,alpha=0.5,label="std") 
    plt.xlabel('Iterations')
    plt.ylabel('Percentage')    
```
